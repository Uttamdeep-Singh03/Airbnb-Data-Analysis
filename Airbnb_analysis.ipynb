## Importing libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("compressed_data.csv")
```

```python
df.head()
```

```python
df.columns
```

# checking missing values

```python
df.isnull().sum()
```

### removing missing values by removing extra unnecessary columns

```python
df.drop(columns=["house_rules","license"],inplace=True,errors="ignore")
```

###### used to ignore errors if occurs due to these columns not present

```python
df.info()
```

#### handling missing values

#### As last review and reviews per month have a lot of null values, so can't remove them as it will reduce our data size 

```python
df["reviews per month"].fillna(0,inplace=True)
```

```python
df["last review"]=pd.to_datetime(df["last review"])
```

```python
df.info()
```

```python
df.fillna({"last review":df["last review"].min()},inplace=True)
```

```python
df.info()
```

```python
df.isnull().sum()
```

#### removing missing values from name and host name

```python
df.dropna(subset=["NAME","host name"],inplace =True)
```

```python
df.head()
```

```python
df["price"]=df["price"].replace(r"[$,]","",regex=True)
df["service fee"]=df["service fee"].replace(r"[$,]","",regex=True)

df["price"]=df["price"].astype(float)
df["service fee"]=df["service fee"].astype(float)
```

```python
df.head()
```

```python
df.info()
```

#### removing duplicates

```python
df.drop_duplicates(inplace=True)
```

#### discriptive statistics

```python
df.describe()
```

## Visualizations

#### What is the distribution of price listings?

```python
plt.figure(figsize=(10,6))
sns.histplot(df["price"],bins=50,kde=True,color="red")
plt.title("Distribution of listing prices")
plt.xlabel("price $")
plt.ylabel("frequency")
plt.show()
```

#### What is distribution of Room type?

```python
plt.figure(figsize=(8,5))
sns.countplot(x="room type",data=df,color="hotpink")
plt.title("Distribution of Room type")
plt.xlabel("Room Type")
plt.ylabel("Frequency")
plt.show()
```

#### Listings distributed across different neighbourhoods

```python
plt.figure(figsize=(12,8))
sns.countplot(y="neighbourhood group", data=df,color="hotpink")
plt.title("Frequency of each neighbourhood")
plt.xlabel("Frequency")
plt.ylabel("different neighbourhoods")
plt.show()
```

#### Relationship between price and room type

```python
plt.figure(figsize=(12,8))
sns.boxplot(x="room type", y="price",hue="room type",data=df,color="hotpink")
plt.xlabel("Room types")
plt.ylabel("Price")
plt.title("Price vs room type")
plt.show()
```

#### No. of reviews changing over time

```python
df["last review"]= pd.to_datetime(df["last review"])
reviews_over_time=df.groupby(df["last review"].dt.to_period("M")).size()
plt.figure(figsize=(12,6))
reviews_over_time.plot(kind="line", color="red")
plt.title("Number of reviews over time")
plt.xlabel("Date")
plt.ylabel("Number of reviews")
plt.show()
```
